```python
geo_df = beMTPL97_filt[['postcode', 'lat', 'long']]
```

```python
unique_locations = geo_df.drop_duplicates().reset_index(drop=True)

# how many unique places 
print(f"Number of unique (postcode, lat, long): {len(unique_locations)}")
```

```python
# first digits in the postcode
unique_locations['postcode_first_digit'] = unique_locations['postcode'].astype(str).str[0]
# how many unique first digits
print(f"Number of unique first digits in postcode: {unique_locations['postcode_first_digit'].nunique()}")
unique_locations['postcode_first_digit'].unique()
```

```python
### Flandes is only the ones starting with 2, 3, 6 and 9...
# lest see the proportion of each first digit
first_digit_counts = unique_locations['postcode_first_digit'].value_counts(normalize=True)
print("Proportion of each first digit in postcode:")
print(first_digit_counts)
```
